```python
"""
НЕ изменяемые объекты:
- Числа
- Строки
- Кортеж
- Булевые значения
- Decimal
- None
"""

from decimal import Decimal
```

```python
hashable_obj_list = [
    123,
    123.123,
    "123",
    (1,2,3),
    True,
    False,
    Decimal("123.123"),
    None,
]

for i in hashable_obj_list:
    print(i, " - ", hash(i))
```

```python
d = {}
for i in hashable_obj_list:
    d[i] = 123

d
```

```python
# ДОбавление в словарь (концепт)

d = {}

key = 123.12
value = "value"
hash_key = hash(123.12)
d[hash_key] = value

d
```

```python
d[hash_key]
```

```python
# Изменяемые объекты нельзя хешировать
"""
Почему ?

Потому что, если изменить изменяемый оюъект, который у нас был бы ключом, т.е 
адресом по которуму лежит значение. То адрес бы изменился и найти 
значение по данному было бы не возможно.
"""
```

```python
l1 = [1,2,3]

# Сделали список ключом

l1= [1, 2]

print(l1)
```

```python
l1 = [1,2,3]
l2 = l1
l2 = [1, 2]

print(l1)
print(l2)
```

```python
l1 = [1,2,3]
l2 = l1.copy()
l2 = [1, 2]

print(l1)
print(l2)
```
