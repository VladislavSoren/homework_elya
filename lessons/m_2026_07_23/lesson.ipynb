```python
for v1, v2 in values:
    print(v1, v2)
```

```python
values = [
    (1,2),
    (11,12),
    (21,22),
]

for v1 in values:
    print(v1)
```

```python
values = [
    (1,2,3),
    (11,12,3),
    (21,22,3),
]

for v1, v2 in values:
    print(v1, v2)
```

```python
values = [
    (1,(2,3)),
    (11,(12,3)),
    (21,(22,3)),
]

for v1, v2 in values:
    print(v1, v2)
```

```python
values = [
    (1,2,3),
    (11,12,3),
    (21,22,3),
]

for v1, v2, v3 in values:
    print(v1, v2, v3)
```
