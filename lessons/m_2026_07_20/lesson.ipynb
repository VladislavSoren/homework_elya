```python
text = "кот сидит на окне кот смотрит в окно кот доволен"
words = text.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

print(word_count)
```

```python
incoming = [("яблоко", 50), ("груша", 30), ("яблоко", 20), ("банан", 40)]
sold = [("яблоко", 35), ("груша", 10), ("банан", 15), ("груша", 5)]

stock = {}

for item, quality in incoming:
    if item in stock:
        stock[item] += quality  
    else:
        stock[item] = quality

for item, quality in sold:
    stock[item] -= quality

print(stock)
```
